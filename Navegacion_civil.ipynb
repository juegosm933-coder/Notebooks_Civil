# EJEMPLO PR√ÅCTICO
No puedo **ver directamente el contenido de ese enlace de ‚Äúshare.google‚Äù** porque estas URLs son atajos de Google que redirigen sin dar acceso al contenido real p√∫blicamente sin m√°s contexto, y no muestran la imagen final al abrirlas desde aqu√≠. ([Share Google][1])

Sin embargo, si lo que quieres es **generar en Google Colab una gr√°fica similar a la de tu ejemplo** (como la convergencia del c√°lculo de œÄ y una visualizaci√≥n), aqu√≠ tienes un c√≥digo completo que autom√°ticamente genera una imagen de salida con estilo profesional que podr√≠as estar buscando:

```python
# üìà Gr√°fica de convergencia para œÄ usando la f√≥rmula de Ramanujan

import math
import matplotlib.pyplot as plt

def pi_ramanujan(n):
    suma = 0
    for k in range(n):
        numerador = math.factorial(4*k) * (1103 + 26390*k)
        denominador = (math.factorial(k)**4) * (396**(4*k))
        suma += numerador / denominador
    
    factor = (2 * math.sqrt(2)) / 9801
    pi_aprox = 1 / (factor * suma)
    return pi_aprox

# N√∫mero de t√©rminos que queremos usar
max_terminos = 10

hallazgos = []
errores = []

for k in range(1, max_terminos + 1):
    aproximacion = pi_ramanujan(k)
    hallazgos.append(aproximacion)
    errores.append(abs(math.pi - aproximacion))

# Mostrar aproximaci√≥n final en consola
print(f"Valor real de œÄ ‚âà {math.pi}")
print(f"Mejor aproximaci√≥n con {max_terminos} t√©rminos ‚âà {hallazgos[-1]:.15f}")
print(f"Error absoluto ‚âà {errores[-1]:.15f}")

# üñºÔ∏è Crear la gr√°fica
plt.figure(figsize=(8, 5))
plt.plot(range(1, max_terminos + 1), errores, marker="o", color="blue")
plt.title("Error absoluto en la aproximaci√≥n de œÄ (Ramanujan)")
plt.xlabel("N√∫mero de t√©rminos")
plt.ylabel("Error absoluto")
plt.grid(True)
plt.savefig("pi_ramanujan_convergencia.png")  # Guarda la imagen localmente
plt.show()
```

1. **Calcula aproximaciones de œÄ** usando la serie de Ramanujan.
2. **Calcula el error absoluto** frente al valor real de œÄ.
